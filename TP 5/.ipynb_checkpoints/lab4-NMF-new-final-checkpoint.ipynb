## NMF

Loss (Regularized Froebnius): $\mathcal{L}(D, S) = \frac{1}{2}\|X-D S\|_{F}^{2}+ \color{red}{\frac{\mu}{2}\|S\|_{F}^{2}+\lambda\|S\|_{1}+\frac{\nu}{2}\|D\|_{F}^{2}}$

Algorithm (multiplicative updates from [Lee and Seung, 1999]: 

\begin{array}{c}
D \leftarrow D \circ \frac{X S^{\top}}{D (S S^{\top} + {\color{red} \nu I_r})} \\
S \leftarrow S \circ \frac{D^{\top} X - {\color{red} \lambda 1_{r \times n}}}{(D^{\top} D + {\color{red} \mu I_r})S}
\end{array}

Fill in the following NMF class:

```python
class MYNMF():
    
    def __init__(self, n_components, random_state=42, max_iter=200):
        self.n_components_ = n_components
        self.max_iter_ = max_iter
        self.components_ = None    # shape (r, p)
        self.random_state = random_state
        self.D_ = None
        self.S_ = None
        self.stop_criteria_  =True
        
    def fit_transform(self, data, mu=0., lambd=0., nu=0.):
        """ Find the parameters mu_ and Sigma_
        that better fit the data
        
        Parameters:
        -----------
        data: (n, p) np.array
               Data matrix
        
        Returns:
        -----
        embedding: (n, r) np.array
                    Embedding matrix        
        """        
        #Initialization with SVD
        #Setting negative values and null values to 1e-3
        n,p = data.shape
        X = data.T
        u, s, vh = np.linalg.svd(X)
        self.D_ = np.where(u<=0, 1e-3,u)[:,:self.n_components_] # D is a (p,p) matrix, we take the first r columns 
        A = np.diag(s)@vh
        self.S_ = np.where(A<=0, 1e-3,A)[:self.n_components_,:]# S is a (n,n) matrix, we take the first r rows
        Ir = np.eye(self.n_components_)
        Irn = np.ones((self.n_components_,n))   
        iteration = 0
        
        #Updating parameters
        while self.stop_criteria_:
            self.D_ = self.D_ * (X@self.S_.T)/(self.D_@(self.S_@self.S_.T + nu*Ir))
            
            self.S_ = self.S_ * (self.D_.T@X-lambd*Irn)/((self.D_.T@self.D_ + mu*Ir)@self.S_)
                
            if iteration>self.max_iter_:
                self.stop_criteria_ = False
            iteration += 1 
        
        #Defining the embedding
        embedding =  self.S_.T #Here S is a (r,n) matrix so we should return its transpose
        
        return embedding  
    
    def transform(self, X,lambd=0.,mu=0.): 
        """To predict unseen data we need to keep D constant and compute the new probability matrix S"""
        D = self.D_.copy()
        n,p = X.shape
        iteration = 0
        Ir = np.eye(self.n_components_)
        Irn = np.ones((self.n_components_,n))
        avg = np.sqrt(X.mean() / self.n_components_)
        S = np.full((self.n_components_, n), avg, dtype=X.dtype)
        while iteration<self.max_iter_:
            S = S * (D.T@X.T-lambd*Irn)/((D.T@D + mu*Ir)@S)
            iteration+=1
            
        return S.T
    
    
    def inverse_transform(self, embedding):
        """ Maps an embedding matrix into the original space. 
        
        Parameters:
        -----------
        embedding : (n, r) np.array
                     Embedding matrix
        
        Returns:
        -----
        data_hat : (n, p) np.array  
                   reconstructed data matrix
        """       
        #Computing the approximation of the data
        data_hat = self.D_@embedding.T
        return data_hat

    def predict_label(self, X):
        """ Predict labels for X : cluster hard-assignment
        
        Parameters:
        -----------
        X: (n, p) np.array
           data matrix
        
        Returns:
        -----
        label assignment : (n, ) np.array
                          Its elements are the predicted classes, from 0 to r-1
        """
        return np.argmax(self.transform(X),axis=1) 
        
        
       
        
    def predict_proba(self, X):
        """ Predict probabilities of different classes for X :  cluster soft-assignment
        
        Parameters:
        -----------
        X: (n, p) np.array
            New data matrix
        
        Returns:
        -----
        label assignment : (n, r) np.array
                          Its elements are the probabilities for different classes
        """
        embedding  = self.transform(X)
        for i in range(n):
            somme = np.sum(embedding[i])
            embedding[i] = embedding[i]/somme
        return embedding #
```

We decided to initialise our NMF algorithm by using a singular value decomposition (SVD). We decided to use this method instead of initialising the two matrices randomly because using the SVD decomposition is like applying principal component analysis to our problem. Thus, we have found a solution to the problem of minimzing the Frobenius norm of (X - DS) where D must be orthogonal and S be defined as H@H.T = diag(a1, a2, ...) with a1, a2, ... the singular values of the matrix S. Now our algorithm will use the solution found for the SVD and look for a solution that follows the following contraints, D and S must have strictly positive coefficients.

## Applications

### First application

For the first application you are going to analyse the following data to give an interpretation of the factorization:

```python
from sklearn.datasets import fetch_olivetti_faces


dataset = fetch_olivetti_faces(shuffle=True)

faces = dataset.data  # n = 400 flattened images of p = 4096 = 64 x 64
image_shape = (64, 64)

n_samples, n_features = faces.shape

# global centering
faces_centered = faces - faces.mean(axis=0)

# local centering
faces_centered -= faces_centered.mean(axis=1).reshape(n_samples, -1)

def plot_gallery(images, image_shape, title='', cmap=plt.cm.gray):
    
    # fit the nb of images into a square
    n_images = len(images)
    nrows = int(np.ceil(np.sqrt(n_images)))
    ncols = nrows
    
    # instantiate figure
    fig, axes = plt.subplots(nrows, ncols, figsize=(2. * ncols, 2.26 * nrows))
    fig.suptitle(title, size=16)
    
    axes_for_display = axes.flatten()[:n_images]
    axes_to_blank = axes.flatten()[n_images:]
    
    for idx, comp in enumerate(images):
        ax = axes_for_display[idx]
        vmax = max(comp.max(), -comp.min())
        ax.imshow(comp.reshape(image_shape), cmap=cmap,
                   interpolation='nearest',
                   vmin=-vmax, vmax=vmax)
        ax.set(xticks=(),
               yticks=())
        
    for ax in axes_to_blank:
        ax.axis('off')

    fig.tight_layout()
    fig.subplots_adjust(0.01, 0.05, 0.99, 0.93, 0.04, 0.)  # insert space below the suptitle
    

plot_gallery(title="Some Faces", 
             images=faces[:25], 
             image_shape=image_shape)
```

Apply *your* NMF algorithm for this dataset and plot the approximated face pictures.

To predict the classes for unseen faces, we only perform the reduction of the dimension to the first 350 faces of the dataset. We will first analyse the results for a reduction to dimension r = 25.

```python
time0 = time.time()
nmf = MYNMF(25)
embedding = nmf.fit_transform(faces[:350]) #for training
W = nmf.D_.T

plot_gallery(title=str(25)+ " classes:", 
             images=W[:25], 
             image_shape=image_shape)
print(time.time()-time0)
```

### Comments on the meaning of W and H

The matrix W is a matrix with representing the features (rows) and classes (columns). It shows the importance of each feature in each class. More precisely, the higher the weight, the more this class is determined by this feature.

The matrix H represents the probability (non normalized) that a sample belongs to a given class. In fact, the higher the value in this matrix, the more this sample is described by this class.

In the above figure, we have 25 pictures that represent the different classes. Given those classes, we will be able to reconstruct every face from the dataset accroding to the probabilities of belonging to a particular class. We can see the importance of every feature in the showed images.

### Reconstructiong the different faces

```python
X_inverse = nmf.inverse_transform(embedding).T
plot_gallery(title="Some Faces", 
             images=X_inverse[:25], 
             image_shape=image_shape)
```

We can see from the above figure that the reconstructed figures look similar to the original dataset. Nonetheless, the reduction of dimension made us loose some information. 
The more we increase the value of r (n_components), the less information we loose. Thus, we should find the optimal value for r. To do so, we tried to compute the evolution of the norm of the difference between the original data X and the approximation DS for different values of r (r being the square of an integer).

### Choice of n_components

```python
r_values = [i**2 for i in range(1,21)]
norm_diff = []
for r in r_values:
    nmf = MYNMF(r)
    embedding = nmf.fit_transform(faces)
    data_hat = nmf.inverse_transform(embedding)
    norm_diff.append(np.linalg.norm(faces.T-data_hat))
plt.title("Variations of the difference of the norm of the data X with its approximation according to the value of the dimension size")    
plt.plot(r_values, norm_diff)
```

According to this graph, the best value for r is the value closest to the origin. In this case we can approximately read that the best value would be r = 64. Let's use this value from now on.

### Choice of the regression parameters

In this section we will use the same notation as in the formula given at the beginning of the lab.

```python
mu_values = [i*0.5 for i in range(1,11)]
norm_diff = []
for i in mu_values:
    nmf = MYNMF(64)
    embedding = nmf.fit_transform(faces, mu = i)
    data_hat = nmf.inverse_transform(embedding)
    norm_diff.append(np.linalg.norm(faces.T-data_hat))

plt.title("Variations of the difference of the norm of the data X with its approximation according to the value of mu")
plt.plot(mu_values, norm_diff)
```

According to this graph, for a chosen value of r and setting to zero all other regularization parameters, the higher the value of mu, the better the approximation. This value corresponds to adding a penalty of the frobenius norm of S (the non normalized probability matrix). 

```python
lambd_values = [i*0.01 for i in range(1,11)]
norm_diff = []
for i in lambd_values:
    nmf = MYNMF(64)
    embedding = nmf.fit_transform(faces, lambd = i)
    data_hat = nmf.inverse_transform(embedding)
    norm_diff.append(np.linalg.norm(faces.T-data_hat))

plt.title("Variations of the difference of the norm of the data X with its approximation according to the value of lambd")
plt.plot(lambd_values, norm_diff)
```

According to this curve, the higher the value for lamba, the worse the approximation. Thus we should keep the value of lambda to zero. Increasing the value of lambda gives more weight to the penalty of the norm-1 of S. This penalty seems to have undesired effects on our approximation.

```python
nu_values = [i*0.5 for i in range(1,11)]
norm_diff = []
for i in nu_values:
    nmf = MYNMF(64)
    embedding = nmf.fit_transform(faces, nu = i)
    data_hat = nmf.inverse_transform(embedding)
    norm_diff.append(np.linalg.norm(faces.T-data_hat))

plt.title("Variations of the difference of the norm of the data X with its approximation according to the value of nu")
plt.plot(lambd_values, norm_diff)
```

Similarly to lambda, increasing nu seems to have negative effects on the approximation we would like to make. Increasing nu adds a penalty of the frobenius norm of D, the weights matrix which seems to be undesired in this case.

#### Conclusion

From the above graphs, it seems that if we want to improve our model, we could add a penalty of the Frobenius norm of the matrix S (the non normalized probability matrix). Thus we could increase the parameter mu.

### Sklearn's NMF

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.datasets import fetch_20newsgroups
from sklearn.decomposition import NMF

n_samples = 2000
n_features = 3000
n_components = 10
n_top_words = 20


def print_top_words(D, feature_names, n_top_words):
    p, r = D.shape
    for topic_idx in range(r):
        topic = D[:, topic_idx]
        message = "Topic #%d: " % topic_idx
        message += " ".join([feature_names[i]
                             for i in topic.argsort()[:-n_top_words - 1:-1]])
        print(message)
    print()


dataset = fetch_20newsgroups(shuffle=True, random_state=1,
                             remove=('headers', 'footers', 'quotes'))
data_samples = dataset.data[:n_samples]


# Use tf-idf features for NMF.
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2,
                                   max_features=n_features,
                                   stop_words='english')
tfidf = tfidf_vectorizer.fit_transform(data_samples).toarray()

# Fit the NMF model, print top words
nmf = NMF(n_components = n_components,solver='mu')
H = nmf.fit_transform(tfidf)
tfidf_feature_names = tfidf_vectorizer.get_feature_names()
# TODO

print_top_words(nmf.components_.T, tfidf_feature_names, n_top_words)
```
